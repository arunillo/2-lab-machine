```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def parse_news(news_count=10):
    news_data = []
    page = 1

    while len(news_data) < news_count:
        url = f"https://lenta.ru/parts/news/{page}/"
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        items = soup.find_all('li', class_='parts-page__item')
        for item in items:
            if len(news_data) >= news_count:
                break

            title_elem = item.find('h3')
            if not title_elem:
                continue

            title = title_elem.text.strip()
            link_elem = item.find('a')
            link = link_elem['href'] if link_elem else ''

            # Безопасное получение даты
            time_elem = item.find('time')
            date = time_elem.get('datetime', '') if time_elem else ''

            # Получаем текст новости
            if link:
                try:
                    article_response = requests.get(f"https://lenta.ru{link}")
                    article_soup = BeautifulSoup(article_response.content, 'html.parser')
                    text_paragraphs = article_soup.find_all('p')
                    text = ' '.join([p.text for p in text_paragraphs])[:300] + '...'
                except:
                    text = ''
            else:
                text = ''

            news_data.append({
                'title': title,
                'date': date,
                'text': text
            })

        page += 1
        if page > 10:  # ограничиваем количество страниц
            break

    return pd.DataFrame(news_data)

# Использование
df = parse_news(5)
print(df)
df.to_csv('news.csv', index=False)
```
